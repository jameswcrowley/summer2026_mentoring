## 2. The Sun: Our Shiny Friend

The sun is our closest star and gives energy for life on Earth! It's a complicated and dynamic object that's super important to study. We'll quickly do a little crash course of the important layers & properties of the Sun!

![alt text](https://scitechdaily.com/images/Extreme-Ultraviolet-Imaging-Telescope.gif)


The Sun is a big ball of Plasma that is constantly producing energy from Nuclear Fusion in its core. This energy makes its way out to the surface of the Sun and eventually reaches the Earth! This energy also indirectly produces space weather and solar storms that can be dangerous for sattelites, astronauts, or GPS (but also produce beautiful Aurora)!


### **I. Solar Core**

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/8/85/Fusion_in_the_Sun.svg/1280px-Fusion_in_the_Sun.svg.png"  style="background-color: #c5dde6;" width="400">

In the solar core, pressure and temperatures are so high that Hydrogen atoms get fused into Helium atoms, producing high-energy photons. These photons heat the Sun and radiate outwards, supporting the Sun against collapsing under its own weight. 

### **II. Radiative Zone**

After photons are produced in the core, they radiate outwards into the *Radiative Zone*

This is a region of the Sun that is very dense - so dense, photons can't make it out without hitting and bouncing off of particles! Photons bounce around in the radiative zone until they eventually make it out of the Radiative Zone - this process is called a "*random walk*".

![alt text](https://images.huffingtonpost.com/2014-06-11-RandomWalk-thumb.gif)

Photons take on average **170,000 years** to bounce around and make it out of the radiative zone!! 

### **III. Convection Zone**

After photons leave the Radiative Zone, the Sun's plasma becomes cooler and more opaque - the photons are absorbed and the plasma is heated. 

The hot plasma starts "convecting": hot plama rises in blobs until it cools down enough to fall back down - think kinda a little like boiling plasma! 

![alt text](https://media3.giphy.com/media/v1.Y2lkPTZjMDliOTUyZ3c5ejlqMHBrNWc5NHFwaWV2OXZtd3lqY2R4NWs2azdqeDU0Mno1ZyZlcD12MV9naWZzX3NlYXJjaCZjdD1n/Xbg9mnADmjOGfq943O/giphy.gif)



### **IV. The Photosphere**

The Photosphere is the top of the convection zone where the plasma becomes transparent! This is the **"surface of the Sun"** where photons to radiate away into space.

![alt text](https://www.sciencealert.com/images/2020-01/solar-cells_1024.gif)

This movie is taken from DKIST over a few minutes. Each convection cell is about the size of Texas! 

### **V. The Chromosphere**

<img src="https://media1.giphy.com/media/v1.Y2lkPTZjMDliOTUyd2FjYWc5d3Yxd2V5NnZmamx3dmVteHZxOHgxYW5pcDZrbXo2NzEwaiZlcD12MV9naWZzX3NlYXJjaCZjdD1n/3o7TKHzhNbSSRl7OGk/200w.gif"  style="background-color: #c5dde6;" width="400">

The chromosphere lies just above the photosphere and appears as a thin, reddish layer during total solar eclipses. It is hotter than the photosphere and is structured by magnetic fields — you can see detailed, quickly evolving tubes of plasma in a variety of sizes and shapes: spicules, filaments, and prominences all trace magnetic activity. Observations in H-alpha and other lines reveal dynamic heating and flows that connect the surface to the outer atmosphere.

### **VI. The Corona**

The corona is the extended outer atmosphere of the Sun. It is extremely hot (millions of kelvin) and low-density. It extends far into space and is the source of the solar wind. Coronal structures such as loops, holes, and coronal mass ejections are shaped by magnetic fields and can be seen during a total solar eclipse! We primarially observe and study the Corona in far-UV and X-ray wavelengths.

![Solar Corona](https://physicsworld.com/wp-content/uploads/2021/09/Judge-frontis-hero.jpg)

## 3. Introduction to DKIST

DKIST (Daniel K. Inouye Solar Telescope) is the world's largest solar telescope - it observes the Sun at REALLY high resolution -it's like a solar microscope. With a 4-meter primary mirror, DKIST observes the Sun at really small scales to study magnetic fields, energy transport, and dynamics.

One of the keys of modern (solar) astronomy is collecting data at different wavelengths.

![alt text](https://upload.wikimedia.org/wikipedia/commons/d/d7/The_sun_in_many_wavelengths.jpg)

A "spectrograph" collects images at different wavelengths. This data is usually saved as 3D "datacubes":

![alt text](https://upload.wikimedia.org/wikipedia/commons/a/ac/Datacube_MUSE_on_NGC_4650A_with_IFU.jpg)

**Main instruments:**
- **Visible Spectropolarimeter (ViSP):** slit-spectrograph measuring spectra in visible wavelengths to study magnetic fields in the photosphere and chromosphere.

![alt text](http://spiff.rit.edu/classes/phys312/workshops/w10b/spectra/slit.gif)

https://nso.edu/telescopes/dkist/instruments/visp/

- **Diffraction-Limited Near-Infrared Spectropolarimeter (DL-NIRSP):** near-infrared integral-field spectropolarimetry for chromospheric and coronal diagnostics at high spatial resolution.

![alt text](https://upload.wikimedia.org/wikipedia/commons/4/4d/Types_of_Integral_Field_Units.png)

https://nso.edu/telescopes/dkist/instruments/dl-nirsp/

- **Cryogenic Near-Infrared Spectropolarimeter (Cryo-NIRSP):** cyrogenically cooled (to -320º F) slit-spectrograph looking for faint coronal infrared lines and large-scale coronal structure measurements.

https://nso.edu/telescopes/dkist/instruments/cryo-nirsp/

- **Visible Broadband Imager (VBI):** high-cadence, high-resolution imaging of photospheric and chromospheric fine structure.

https://nso.edu/telescopes/dkist/instruments/vbi/

Together these instruments let DKIST probe magnetic fields and structures from the photosphere up through the chromosphere and into the corona.
