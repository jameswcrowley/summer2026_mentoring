Version control helps you save, organize, and share changes to your files. This notebook covers the basic ideas: commits, branches, pushing, and pulling.


## Commits
- A commit is a saved snapshot of your project.
- Each commit has a message describing what changed.
- Think of commits as milestones in your work.

## Branches
- A branch is a separate line of work.
- Use branches to try features without changing the main project.
- Common branch names are `main`, `develop`, and feature names.

<img src="https://gcapes.github.io/git-course/fig/feature-branches.svg"  style="background-color: #c5dde6;" width="800">


## Pushing and Pulling
- `push` sends your commits to a remote repository.
- `pull` brings changes from the remote repository to your local copy.
- These commands keep your work synced with others


## Common Workflow
1. Create or switch to a branch.
2. Make changes and commit them.
3. Push the branch to the remote repository.
4. Pull updates from the remote before you start new work.

## Key Commands
- `git init` — start version control in a folder.
- `git add` — stage files for a commit.
- `git commit -m "message"` — save a snapshot.
- `git branch` — list or create branches.
- `git checkout` / `git switch` — move between branches.
- `git push` — upload commits to a remote.
- `git pull` — download commits from a remote.
