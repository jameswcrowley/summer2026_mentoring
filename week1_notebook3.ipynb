**Mutable and Immutable**

Numbers are immutable.  Lists are mutable, meaning changeable.

Write x=5

The 5 is immutable, because the nature of an integer cannot be changes.

*Pointers*

 A pointer is a chunk of memory of fixed size.
 
 Within that chunk of memory Python saves all the information relevant.\
 It remembers where the first byte is in memory.\
           It remembers the number of bytes in the answer.\
           It remembers the type of variable.
 
When the expression x=5 is encountered, it first strips the unnecessary spaces.\
Then it evaluates the expression to the right of the equal sign.  \
It recognizes the expression as a two-byte immutable integer (5). 

It then looks at the left side of the equals sign. \
It finds a variable name with the string value "x". \
Somewhere in memory Python has created a list of names of defined variables.\
Think of this as a two column table. The first column has the name of the variables and the second column has the memory position of the pointer that is associated with that variable name.

If x is already defined it finds the pointer associated with x\
if undefined it creates the name and pointer associated with the name.  \
Then it records in the pointer that the value to be stored is a single integer.  \
It saves the value 5 somewhere in computer memory and records in the pointer where that value resides.

Then Python goes to the next line, where it encounters  x=x+2.  \
When Python encounters the "x" on the right side, it goes down the list to find the pointer to the value of "x".  \
In our example, it points to an integer 5. \
It identifies the 2 as an integer and creates a pointer to that. \
It identifies the + as an operator command to add the contents of the two pointers.  \
It identifies + as an operator.  \
Somewhere there is an addition function.  \
It sends the two pointers to that function, which returns a single pointer containing the value of the sum.  \
That pointer replaces the pointer that was associated with x.

Consider the case of a list.

```python
mylist=["The first four primes are:",1,2,3,5]
print(mylist[3])
```

What happened here?  
When working on the right side of the assignment statement (the expression), Python encountered a [ square bracket, and further along a closing square bracket ].  
That told Python that it was to create a pointer to a "list" variable and associate that pointer (target) to a list with the name "mylist".

Inside the brackets Python looks for commas to separate the elements of the list. \
Each of those elements will have a pointer.  \
So the "mylist" will be associated with a pointer to a series of pointers. \
The "mylist" pointer will contain the number of pointers in the list.

*Python, folks, is all about lists of pointers.  \
And pointers to pointers.  
Keep this in mind and the syntax of Python will start to make sense.*

It's incredibly elegant when you get to know it.

Notice that the print statement contains mylist[3]. \
That, of course, means print only the fourth element in the mylist list.  \
So what python does is open up the mylist pointer which points to a list of pointers.  \
It finds the fourth pointer in the list, which, in this case, is a pointer to a single integer. \
Saved in the memory element to which it points is the value 3

mylist is associated with a pointer  

that pointer contains the information of the location in memory where the first of the five pointers reside.  

Since all the pointers are the same size, the fourth value can be accessed by adding the size of three more pointers to the position of the first pointer.  

The print(mylist[3]) statement is fed only the fourth pointer.  \
It doesnt need to know about the other pointers. 
So print() then prints the information that is pointed to by the pointer it is given. \
In this case an integer of value 3.
