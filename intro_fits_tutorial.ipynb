```python
# Accessing the image data:
test_fits_data = fits.open('/Users/jamescrowley/Documents/summer_2026/research/pid_4_62/IHFDSO/VISP_2026_05_13T23_56_31_823023_00589100_I_IHFDSO_L1.fits')[1].data

# Accessing the fits header (metadata):
test_fits_header = fits.open('/Users/jamescrowley/Documents/summer_2026/research/pid_4_62/IHFDSO/VISP_2026_05_13T23_56_31_823023_00589100_I_IHFDSO_L1.fits')[1].header
```

```python
# The shape of the data:
print(test_fits_data.shape)
```

```python
# Accesssing one of the FITS header keywords:
print(test_fits_header['CUNIT3'])
```

```python
plt.figure(figsize = [20, 8])

# Plotting a spectra from a single pixel: 
plt.subplot(1,2,1)
plt.plot(test_fits_data[0, :, 10])
plt.axvline(10, c = 'r', ls = '--')

# Plotting the average of the first 20 wavelength indexes across ALL pixels across the slit
plt.subplot(1,2,2)
plt.plot(test_fits_data[0, 10, :])
plt.plot(np.nanmean(test_fits_data[0, :20, :], axis = 0))
plt.ylim(0.8, 1.2)
```
